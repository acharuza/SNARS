```python
from p_8_1 import clustering_coefficient, compute_Ck, plot_Ck
import networkx as nx
from scipy.io import mmread
import matplotlib.pyplot as plt
```

```python
def read_mtx(filename):
    A = mmread(filename)
    G = nx.from_scipy_sparse_array(A)
    return G
```

```python
econ_poli = read_mtx('econ-poli.mtx') # it was in economy networks on network repository
C = clustering_coefficient(econ_poli)
k_vals, Ck_vals = compute_Ck(econ_poli, C)
plot_Ck(k_vals, Ck_vals, "C(k) vs k for econ-poli network")
```

```python
power_US_grid = read_mtx('power-US-grid.mtx') # it was in power grid networks on network repository
C = clustering_coefficient(power_US_grid)
k_vals, Ck_vals = compute_Ck(power_US_grid, C)
plot_Ck(k_vals, Ck_vals, "C(k) vs k for power-US-grid network")
```
